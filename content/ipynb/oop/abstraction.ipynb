# Моделирование

## Сокрытие информации (information hiding) и инкапсуляция (encapsulation)
**Сокрытие информации** -- процесс сокрытия реализации или функциональных деталей объекта соответствующим образом.


Это также иногда называют инкапсуляцией, но на самом деле инкапсуляция является более всеобъемлющим термином. Инкапсулированные данные не обязательно скрыты. **Инкапсуляция** -- это буквально создание капсулы -- можно представить создании капсулы времени. Если вы помещаете кучу информации в капсулу времени, закрываете ее и закапываете, она одновременно инкапсулирована, а информация скрыта. С другой стороны, если капсула времени не была похоронена и раскрыта или сделана из прозрачного пластика, предметы внутри нее все еще инкапсулированы, но информация не скрывается.

Различие между инкапсуляцией и сокрытием информации в значительной степени не имеет значения, особенно на уровне проектирования. Многие практические ссылки используют эти термины взаимозаменяемо. Как программисты Python, мы фактически не имеем или не нуждаемся в истинном сокрытии информации поэтому более всеобъемлющее определение для инкапсуляции подходит.

## Публичный интерфейс (API)
Публинчный интерфейс -- это тот набор методов и данных, который подразумевается для использование из-вне. Публичный интерфейс очень важен. Он должен быть тщательно разработан, так как его трудно изменить в будущем. Изменение интерфейса требует изменение всех объектов, использующих его.

## Абстрагирование 

Объекты в программировании хоть порой и отражают какие-то реальные вещи, но разумеется ими не являются. Объекты -- это модели. Модель -- это абстракция реального понятия.

**Абстракция** -- это еще одно объектно-ориентированное понятие, связанное с инкапсуляцией и сокрытием информации. Проще говоря, абстракция означает работу с уровнем детализации, наиболее подходящим для данной задачи.

Сжимая весь этот жаргон в пару предложений: абстракция - это процесс инкапсуляции информации с помощью публичных и приватных интерфейсов. 


На практике объекты в ООП иногда представляют объекты реального мира, например, людей или книги. Но намного более часто, они отражают какие-то абстрактные идеи.
